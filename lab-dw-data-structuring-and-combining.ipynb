```python
# Your code goes here
import pandas as pd

df1 = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv")
df2 = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv")
df3 = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv")

df1.dropna(how='all', inplace=True)
df2.dropna(how='all', inplace=True)
df3.dropna(how='all', inplace=True)

rename_dict = {'ST': 'State', 'GENDER': 'Gender'}

def rename_cols(df, rename_dict):
    existing_columns = {col: rename_dict[col] for col in df.columns if col in rename_dict}
    df2 = df.rename(columns=existing_columns)
    return df2


df1["Number of Open Complaints"] = df1["Number of Open Complaints"].apply(lambda x: int(x.split("/")[1]))
df2["Number of Open Complaints"] = df2["Number of Open Complaints"].apply(lambda x: int(x.split("/")[1]))

df1 = rename_cols(df1, rename_dict)
df2 = rename_cols(df2, rename_dict)
df3 = rename_cols(df3, rename_dict)

cols = list(df1.columns)

df1 = df1[cols]
df2 = df2[cols]
df3 = df3[cols]
```

```python
display(df1.head())
display(df2.head())
display(df3.head())
```

```python

total_df = pd.concat([df1, df2,df3], axis=0)

total_df.dropna(how='all', inplace=True)

total_df.head()
```

```python
def clean_cols(df):
    clean_df = df.copy()
    
    clean_df.columns = [ item.lower().replace(" ","_") for item in df.columns ]
    
    clean_df.rename(columns= {'st': 'state'}, inplace=True)
    
    clean_df.columns
    return(clean_df)

total_df = clean_cols(total_df)

total_df.head()
```

```python
total_df["gender"].unique()
```

```python
def clean_gender(df):

    clean_df = df.copy()
    gender_mapping = {"F": "F", "M": "M", "Femal": "F", "Male": "M", "female": "F" , "" : ""}   
    clean_df["gender"] = df["gender"].map(gender_mapping)   
    return clean_df

total_df = clean_gender(total_df)
```

```python
def clean_state(df):
    clean_df = df.copy()
    state_mapping = {"Washington": "Washington", "Arizona": "Arizona", "Nevada": "Nevada", "California": "California", "Oregon": "Oregon" , "Cali" : "California", 'AZ': "Arizona", 'WA' : "Washington"}   
    clean_df["state"] = df["state"].map(state_mapping)   
    return clean_df

total_df = clean_state(total_df)
```

```python
def clean_edu(df):
    clean_df = df.copy()
    edu_mapping = {"Bachelors": "Bachelor", "Master": "Master", "Bachelor": "Bachelor",
                     "High School or Below": "High School or Below", "College": "College" ,
                     "Doctor" : "Doctor"}   


    clean_df["education"] = df["education"].map(edu_mapping)   
    return clean_df

clean_df = clean_edu(total_df)

clean_df.head()
```

```python
def clean_CLV(df):
    clean_df = df.copy()

    # Define a function to clean individual values
    def clean_value(value):
        if isinstance(value, str) and '%' in value:
            return 100 * float(value.replace('%', ''))
        return value  # Return the value unchanged if no '%' is present

    # Apply the cleaning function to the column
    clean_df["customer_lifetime_value"] = clean_df["customer_lifetime_value"].apply(clean_value)
    
    return clean_df

display(clean_df["customer_lifetime_value"].isna().sum())

total_clean_df = clean_CLV(clean_df)

display(total_clean_df.isna().sum())
```

```python
def clean_car(df):
    clean_df = df.copy()
    clean_df["vehicle_class"] = clean_df["vehicle_class"].replace("Sports Car", "Luxury")
    clean_df["vehicle_class"] = clean_df["vehicle_class"].replace("Luxury SUV", "Luxury")
    clean_df["vehicle_class"] = clean_df["vehicle_class"].replace("Luxury Car", "Luxury")   
    
    return clean_df

my_df = clean_car(total_clean_df)
```

```python
display(my_df.isna().sum())
display(my_df.shape)
```

```python
my_df = my_df.dropna()

display(my_df.isna().sum())
display(my_df.shape)
```

# Challenge 2: Structuring Data

In this challenge, we will continue to work with customer data from an insurance company, but we will use a dataset with more columns, called marketing_customer_analysis.csv, which can be found at the following link:

https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv

This dataset contains information such as customer demographics, policy details, vehicle information, and the customer's response to the last marketing campaign. Our goal is to explore and analyze this data by performing data cleaning, formatting, and structuring.

```python
import pandas as pd

marketing_url = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv"

mkt_df = pd.read_csv(marketing_url)

mkt_df.head()
```

```python
mkt_df.columns
```

1. You work at the marketing department and you want to know which sales channel brought the most sales in terms of total revenue. Using pivot, create a summary table showing the total revenue for each sales channel (branch, call center, web, and mail).
Round the total revenue to 2 decimal points.  Analyze the resulting table to draw insights.

```python
PT = mkt_df.pivot_table(index='sales_channel', columns='policy_type', values=['customer_lifetime_value'], aggfunc="sum").round(2)

PT["Total"] = PT.sum(axis=1)

PT

#Agents bring in the most recenue across all policy types
```

```python
PT2 = mkt_df.pivot_table(index='sales_channel', columns='gender', values=['customer_lifetime_value'], aggfunc="sum").round(2)

PT2["Total"] = PT2.sum(axis=1)

PT2
```

2. Create a pivot table that shows the average customer lifetime value per gender and education level. Analyze the resulting table to draw insights.

```python
PT3 = mkt_df.pivot_table(index='gender', columns='education', values=['customer_lifetime_value'], aggfunc="mean").round(2)

PT3["Total"] = PT3.mean(axis=1)

PT3
```
